
# Plot 2D data on 3D plot


Demonstrates using ax.plot's zdir keyword to plot 2D data on
selective axes of a 3D plot.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from pylab import rcParams
rcParams['figure.figsize'] =10, 7

fig = plt.figure()
# ax = fig.gca(projection='3d')
ax = fig.add_subplot(111, projection='3d')

# where to evaluate the densities
x = np.linspace(-2, 12, 100)

def gaussian(x, mu, sigma):
    return 1./(np.sqrt(2.*np.pi)*sigma)*np.exp(-np.power((x - mu)/sigma, 2.)/2)


c1 = gaussian(x, 6, 1)
c2 = gaussian(x, 6.5, 1)
c3 = gaussian(x, 7, 1)
sum_y1 = (1/5)*(gaussian(x, 2, 1)+gaussian(x, 2.5, 1)+gaussian(x, 3, 1)+gaussian(x, 1, 1)+gaussian(x, 6, 1))
sum_y2 = (1/3)*(c1 + c2 + c3)

plt.plot(x, sum_y1, zs=1, zdir='y',label=r'$P(X|\theta=1)$')
plt.plot(x, sum_y2, zs=2, zdir='y',label=r'$P(X|\theta=2)$')

# Make legend, set axes limits and labels
ax.legend()
ax.set_xlim(-2., 12.)
ax.set_ylim(.5, 2.5)
ax.set_zlim(0, .4)
# plt.fill_between(x, 0, sum_y1, zdir='y')
# plt.fill_between(x, 0, sum_y2)
ax.set_xlabel(r'$X$',fontsize=20)
ax.set_ylabel(r'$\theta$',fontsize=20)
ax.set_zlabel(r'$f_{X|\theta}(x|\theta)$',fontsize=20)
ax.set_title(r'Joint PDF $f_{X|\theta}(x|\theta)$',fontsize=20)
# ax.set_title(r'Joint PDF of X and $\theta$')
# verts = [(x[i],z[i],y[i]) for i in range(len(x))] + [(x.max(),0,0),(x.min(),0,0)]
# ax.add_collection3d(Poly3DCollection([verts])) # Add a polygon instead of fill_between
ax.legend(prop={'size': 10})
# Customize the view angle so it's easier to see that the scatter points lie
# on the plane y=0
ax.view_init(elev=30., azim=-60)

plt.show()
```

```python
def polygon_under_graph(xlist, ylist):
    """
    Construct the vertex list which defines the polygon filling the space under
    the (xlist, ylist) line graph.  Assumes the xs are in ascending order.
    """
    return [(xlist[0], 0.), *zip(xlist, ylist), (xlist[-1], 0.)]

# https://matplotlib.org/3.1.1/gallery/mplot3d/polys3d.html
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
rcParams['figure.figsize'] =10, 7

# define your distributions
d1 = stats.norm(4, 2)
d2 = stats.norm(7, 0.5)

# set mixture component weights
mc = [0.5, 0.5]
mc = mc / np.sum(mc) # ensuring they sum to 1

# where to evaluate the densities
x = np.linspace(-2, 12, 100)
# calculate density and apply mixture weights
c1 = d1.pdf(x) * mc[0]
c2 = d2.pdf(x) * mc[1]


def gaussian(x, mu, sigma):
    return 1./(np.sqrt(2.*np.pi)*sigma)*np.exp(-np.power((x - mu)/sigma, 2.)/2)
c1 = gaussian(x, 6, 1)
c2 = gaussian(x, 6.5, 1)
c3 = gaussian(x, 7, 1)
sum_y1 = (1/5)*(gaussian(x, 2, 1)+gaussian(x, 2.5, 1)+gaussian(x, 3, 1)+gaussian(x, 1, 1)+gaussian(x, 6, 1))
sum_y2 = (1/3)*(c1 + c2 + c3)

# plot everything
plt.plot(x, sum_y1, label=r'$P(X|\theta=1)$')
# plt.plot(x, c2, label='Component 2')
plt.plot(x,sum_y2, label=r'$P(X|\theta=2)$')
# plt.axvline(4.5, c='k')
# plt.fill_between(x, 0, sum_y1)
# plt.fill_between(x, 0, sum_y1, where=x>=4.5)
# plt.fill_between(x, 0, sum_y2, where=x<=4.64)
plt.xlabel(r'$X$',fontsize=20)
plt.ylabel(r'$f_{X|\theta}(x|\theta)$',fontsize=20)
# plt.title(r'Joint PDF $f_{X|\theta}(x|\theta)$',fontsize=20)
# plt.text(-1, 0.25, r'$P(error|\theta=2)$', fontsize=15)
# plt.text(8, 0.25, r'$P(error|\theta=1)$', fontsize=15)
plt.title(r'Parzen-window',fontsize=20)
plt.legend(prop={'size': 15})
plt.ylim(0, 0.4)
plt.show()
```
